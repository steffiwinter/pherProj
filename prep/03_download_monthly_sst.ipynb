Get GLORYS data from: https://data.marine.copernicus.eu/product/GLOBAL_MULTIYEAR_PHY_ENS_001_031/services

```python
ds = copernicusmarine.open_dataset(dataset_id="cmems_mod_glo_phy-all_my_0.25deg_P1M-m")
ds
```

```python
sst = ds.thetao_glor.isel(depth=0)
sst = sst.coarsen(longitude=2, latitude=2, boundary="trim").mean()
sst
```

```python
sst.to_dataset(name="sea_surface_temperature").drop_encoding().to_netcdf("../../data/sst_1m.nc", format="NETCDF3_64BIT")
```
